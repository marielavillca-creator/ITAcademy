### 1. Consulta la API pública JSONPlaceholder utilizando el método GET para obtener:

Lista de publicaciones (/posts) , Lista de usuarios (/users) , Lista de tareas (/todos)



```python
import requests
import pandas as pd

# API DE LABORATORIO
base_url = "https://jsonplaceholder.typicode.com"
```

```python
url = "https://jsonplaceholder.typicode.com/posts"

respuesta = requests.get(url)

posts = respuesta.json()

print("Código de estado:", respuesta.status_code)
print("Cantidad de publicaciones:", len(posts))
```

```python
url = "https://jsonplaceholder.typicode.com/users"

respuesta = requests.get(url)

usuarios = respuesta.json()

print("Código de estado:", respuesta.status_code)
print("Cantidad de usuarios:", len(usuarios))
```

```python
url = "https://jsonplaceholder.typicode.com/todos"

respuesta = requests.get(url)

tareas = respuesta.json()

print("Código de estado:", respuesta.status_code)
print("Cantidad de tareas:", len(tareas))
```

### 2. Muestra en pantalla:

La cantidad total de cada recurso.
El código de estado de cada petición.

```python
recursos = ["posts", "users", "todos"]

for recurs in recursos:

    resposta = requests.get(f"{base_url}/{recurs}")
    
    dades = resposta.json()
    
    print(f"Recurs: /{recurs}")
    print(f"  - Quantitat total: {len(dades)}")
    print(f"  - Codi d'estat: {resposta.status_code}")
    print("-" * 30)
```

El título.
El cuerpo del mensaje.

Muestra:

### 5. Realiza una petición PATCH para modificar parcialmente una publicación existente.

Muestra:

La respuesta en formato JSON.
El código de estado.


```python
url = "https://jsonplaceholder.typicode.com/posts/1"

datos_actualizados = {
    "title": "Título actualizado con PATCH"
}

respuesta = requests.patch(url, json=datos_actualizados)

print("Código de estado:", respuesta.status_code)
print("Respuesta JSON:")
print(respuesta.json())
```

### 6. Realiza una petición DELETE sobre una publicación.

Muestra:

La respuesta en formato JSON.
El código de estado.

```python
url = "https://jsonplaceholder.typicode.com/posts/1"

respuesta = requests.delete(url)

print("Código de estado:", respuesta.status_code)
print("Respuesta:")
print(respuesta.text)
```

### 3. Realiza una petición GET sencilla.

Muestra: El código de estado de la respuesta.
Imprime de forma clara algunos de los campos de la respuesta JSON.


```python
import requests

url = "https://dog.ceo/api/breeds/list/all"

respuesta = requests.get(url)

print("Código de estado:", respuesta.status_code)
```

```python
datos = respuesta.json()

print("Estado:", datos["status"])

print()

print("Primeras razas:")

contador = 0

for raza in datos["message"]:
    print(raza)
    contador += 1

    if contador == 10:
        break
```

### 4. Convierte los datos a un DataFrame de pandas.

Muestra: Las primeras filas del DataFrame.

```python
import pandas as pd

lista_razas = []

for raza in datos["message"]:
    lista_razas.append(raza)

df = pd.DataFrame(lista_razas, columns=["Raza"])

df.head()
```

```python
df.to_csv("razas_perros.csv", index=False)

print("Archivo guardado correctamente.")
```
