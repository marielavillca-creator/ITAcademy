### Ejercicio 1 Calculadora del Índice de Masa Corporal (IMC)
Se desarrolla una función que calcula el IMC a partir del peso y la altura de una persona, además de clasificar el resultado según las categorías habituales.

```python
def calcular_imc(peso, altura):
    """
    Calcula el Índice de Masa Corporal (IMC) y devuelve
    la categoría correspondiente según el resultado.
    """

    imc = peso / (altura ** 2)

    if imc < 18.5:
        categoria = "Bajo peso"
    elif imc < 25:
        categoria = "Peso normal"
    elif imc < 30:
        categoria = "Sobrepeso"
    else:
        categoria = "Obesidad"

    return round(imc, 2), categoria
```

```python
def pedir_numero(mensaje):
    """
    Solicita al usuario un número positivo y valida
    que la entrada sea correcta.
    """

    while True:
        try:
            valor = float(input(mensaje).replace(",", "."))
            if valor > 0:
                return valor
            print("El valor debe ser mayor que 0.")
        except ValueError:
            print("Debes introducir un número válido.")
```

```python
def principal_imc():
    peso = pedir_numero("Introduce tu peso en kilogramos: ")
    altura = pedir_numero("Introduce tu altura en metros: ")

    imc, categoria = calcular_imc(peso, altura)

    print(f"\nTu IMC es {imc}")
    print(f"Clasificación: {categoria}")
```

```python
principal_imc()
```

### Ejercicio 2 Convertidor de temperaturas

Existen diversas unidades de temperatura utilizadas en diferentes contextos y regiones. Las más comunes son Celsius (°C), Fahrenheit (°F) y Kelvin (K). También existen otras unidades como Rankine (°Ra) y Réaumur (°Re).
Selecciona al menos 2 conversores, de manera que al introducir una temperatura devuelva, como mínimo, dos conversiones, de forma que se puedan guardar (recuerda que un print() nunca puede guardarse).
Consejo: Intenta validar previamente los datos para que envíe un mensaje de advertencia si los datos introducidos por el usuario no tienen el formato adecuado.
(EXTRA): Piensa en una forma de almacenar todas las posibles conversiones en un único objeto (¿Lista? ¿Diccionario? ¿DataFrame?) en lugar de escribir muchos if-else según la temperatura de origen y destino.

```python
def convertir_temperatura(valor, unidad):
    """
    Convierte una temperatura entre Celsius,
    Fahrenheit y Kelvin.
    """

    conversiones = {
        "C": {
            "Celsius": valor,
            "Fahrenheit": (valor * 9 / 5) + 32,
            "Kelvin": valor + 273.15
        },
        "F": {
            "Celsius": (valor - 32) * 5 / 9,
            "Fahrenheit": valor,
            "Kelvin": (valor - 32) * 5 / 9 + 273.15
        },
        "K": {
            "Celsius": valor - 273.15,
            "Fahrenheit": (valor - 273.15) * 9 / 5 + 32,
            "Kelvin": valor
        }
    }

    return conversiones.get(unidad.upper())
```

```python
def es_float(valor):
    """
    Comprueba si un valor puede convertirse a float.
    """
    try:
        float(valor)
        return True
    except:
        return False
```

```python
def principal_temperatura():
    valor = input("Introduce la temperatura: ")
    unidad = input("Introduce unidad (C/F/K): ")

    if not es_float(valor):
        print("Error: valor no numérico")
        return

    resultado = convertir_temperatura(float(valor), unidad)

    if not resultado:
        print("Error: unidad inválida")
        return

    for k, v in resultado.items():
        print(f"{k}: {v:.2f}")
```

### Ejercicio 3 Contador de palabras 
Escribe una función que, dado un texto, muestre cuántas veces aparece cada palabra. Intenta gestionar todas las casuísticas posibles que puedan hacer que el programa no funcione correctamente.
(EXTRA): ¿Cuál es la longitud media de las palabras del texto que has escrito?

```python
def limpiar_texto(texto):
    """
    Limpia un texto eliminando signos de puntuación
    y convirtiendo todo a minúsculas.
    """

    texto = texto.lower()
    texto = re.sub(r"[^a-záéíóúñü\s]", "", texto)
    return texto.split()
```

```python
def contar_palabras(texto):
    """
    Cuenta la frecuencia de palabras en un texto.
    """
    return limpiar_texto(texto)
```

```python
def frecuencia_palabras(palabras):
    return dict(Counter(palabras))
```

```python
def longitud_media_palabras(palabras):
    return mean(map(len, palabras))
```

```python
def principal_texto():
    texto = input("Introduce un texto: ")

    palabras = contar_palabras(texto)

    print("\nFrecuencia de palabras:")
    print(frecuencia_palabras(palabras))

    print("\nLongitud media:")
    print(round(longitud_media_palabras(palabras), 2))
```

```python
principal_texto()
```

### 4. Diccionario inverso (con posibilidad de duplicados)
Resulta que el cliente tiene una encuesta muy antigua que se almacena en un diccionario y necesita los resultados al revés, es decir, intercambiando las claves y los valores.
Las claves y los valores en el diccionario original deben ser únicos. Si este no es el caso, la función deberá imprimir un mensaje de advertencia junto con una lista de los valores asociados a la clave repetida.
Observa los siguientes ejemplos para ver la salida esperada.

```python
def invertir_diccionario(diccionario):
    """
    Invierte las claves y valores de un diccionario.
    """

    invertido = {}

    for k, v in diccionario.items():

        if v not in invertido:
            invertido[v] = k
        else:
            if not isinstance(invertido[v], list):
                invertido[v] = [invertido[v]]
            invertido[v].append(k)

    return invertido
```

```python
def principal_diccionario():

    datos = {
        "X": "apple",
        "Y": "banana",
        "Z": "banana",
        "F": "apple"
    }

    print(invertir_diccionario(datos))
```

### 2. Conversión de tipos de datos
El cliente recibe una lista de datos y necesita generar dos listas:
•	La primera contendrá todos los elementos que se pudieron convertir a tipo float. 
•	La segunda contendrá los elementos que no se pudieron convertir. 
Ejemplo de lista recibida:
['1.3', 'one', '1e10', 'seven', '3-1/2',
 ('2', 1, 1.4, 'not-a-number'),
 [1, 2, '3', '3.4']]
Imagen 4  
['1.3', 'one', 'le10', 'seven', '3-1/2', ('2',1,1.4,'not-a-number'), [1,2,'3','3.4']]

conversion([ '1.3', 'one' 'le10', 'seven', '3-1/2', ('2',1,1.4, 'not-a-number'), [1,2,'3', '3.4']])

([1.3, 10000000000.0, 2.0, 1.0, 1.4, 1.0, 2.0, 3.0, 3.4], ['one', 'seven', '3-1/2', 'not-a-number'])


```python
def es_float(valor):
    """
    Comprueba si un valor puede convertirse a float.

    Parameters
    ----------
    valor : any

    Returns
    -------
    bool
        True si puede convertirse a float.
    """

    try:
        float(valor)
        return True

    except (ValueError, TypeError):
        return False
    
def conversion(lista):
    """
    Separa los elementos convertibles
    y no convertibles a float.

    Parameters
    ----------
    lista : list

    Returns
    -------
    tuple
        (convertibles, no_convertibles)
    """

    convertibles = []
    no_convertibles = []

    for elemento in lista:

        if isinstance(elemento, (list, tuple)):

            for subelemento in elemento:

                if es_float(subelemento):
                    convertibles.append(float(subelemento))
                else:
                    no_convertibles.append(subelemento)

        else:

            if es_float(elemento):
                convertibles.append(float(elemento))
            else:
                no_convertibles.append(elemento)

    return convertibles, no_convertibles
```

```python
datos = [
    '1.3',
    'one',
    '1e10',
    'seven',
    '3-1/2',
    ('2', 1, 1.4, 'not-a-number'),
    [1, 2, '3', '3.4']
]

validos, invalidos = conversion(datos)

print("Convertibles:")
print(validos)

print("\nNo convertibles:")
print(invalidos)
```

En este punto, el cliente ha detectado un problema con las contraseñas que utilizan sus trabajadores:

fechas de cumpleaños o similares.
Para solucionarlo, nos ha encargado una función de Python que genere contraseñas más seguras.
La función debe depender de los siguientes parámetros:
longitud (int): longitud de la contraseña
mayusculas (bool = True): si deben aparecer mayúsculas
minusculas (bool = True): si deben aparecer minúsculas
numeros (bool = True): si deben aparecer números
signos (bool = False): si deben aparecer caracteres especiales (, - $ ? o similares)
Por lo tanto, si ejecutamos la función de la siguiente forma:
crear_contrasena(10, True, True, True, True)
Deberíamos obtener una salida (que debemos poder guardar) del estilo:
9Er,5Vn8P$
Asegúrate de que se cumplan todos los criterios y de que las contraseñas sean realmente aleatorias.
(EXTRA): Explora cómo podríamos hacer que la función copiara automáticamente la contraseña al portapapeles del ordenador (como si la hubiéramos seleccionado y presionado Ctrl+C).


```python
def crear_contrasena(longitud,
                     mayusculas=True,
                     minusculas=True,
                     numeros=True,
                     signos=False):

    mayus = string.ascii_uppercase
    minus = string.ascii_lowercase
    nums = string.digits
    signos_lista = ",-$.!?@#%&*"

    grupos = []
    pool = []

    if mayusculas:
        grupos.append(list(mayus))
        pool += list(mayus)

    if minusculas:
        grupos.append(list(minus))
        pool += list(minus)

    if numeros:
        grupos.append(list(nums))
        pool += list(nums)

    if signos:
        grupos.append(list(signos_lista))
        pool += list(signos_lista)

    if len(pool) == 0:
        return "Error"

    if longitud < len(grupos):
        return "Error: longitud insuficiente"

    password = [random.choice(g) for g in grupos]
    password += list(np.random.choice(pool, longitud - len(password)))
    random.shuffle(password)

    return "".join(password)
```

```python
def copiar_portapapeles(texto): # extra
    """
    Copia un texto al portapapeles.
    """
    pyperclip.copy(texto)
```

```python
def principal_password():

    pwd = crear_contrasena(10, True, True, True, True)

    print("Contraseña:", pwd)

    copiar_portapapeles(pwd)
```

### 2. Procesamiento simple de datos
Una compañera de trabajo nos ha pedido un favor, aprovechando que sabe que estamos aprendiendo a programar.
Tiene un histórico de partidos de fútbol catalán en un archivo donde se almacenan los nombres de los equipos y los resultados. Necesita que procesemos los datos de forma automática para extraer la información que necesita.
Utiliza el archivo:
historic_partits.txt
Necesita un programa que devuelva:
•	El número total de goles marcados por cada equipo. 
•	El nombre del equipo más goleador. 
•	El nombre del equipo más goleado. 
•	La clasificación global (victoria = 3 puntos, empate = 1 punto, derrota = 0 puntos). 
Como referencia, los equipos con más puntos serían los siguientes:
Imagen resultado partidos
Puedes utilizar el formato que te resulte más conveniente.


```python
def leer_partidos(ruta):

    partidos = []

    with open(ruta, encoding="utf-8") as f:
        for linea in f:
            if linea.strip():
                local, res, visitante = linea.split("\t")
                gl, gv = map(int, res.split("-"))
                partidos.append((local, gl, gv, visitante))

    return partidos
```

```python
def procesar(partidos):

    gf = defaultdict(int)
    gc = defaultdict(int)
    puntos = defaultdict(int)

    for l, gl, gv, v in partidos:

        gf[l] += gl
        gf[v] += gv

        gc[l] += gv
        gc[v] += gl

        if gl > gv:
            puntos[l] += 3
        elif gv > gl:
            puntos[v] += 3
        else:
            puntos[l] += 1
            puntos[v] += 1

    return gf, gc, puntos
```

```python
def maximo(dic):
    return max(dic, key=dic.get)
```

```python
def principal_partidos():

    ruta = "historic partits.txt"

    partidos = leer_partidos(ruta)

    gf, gc, puntos = procesar(partidos)

    print("Goles a favor:", dict(gf))
    print("Más goleador:", maximo(gf))
    print("Más goleado:", maximo(gc))

    print("\nClasificación:")
    for eq, pts in sorted(puntos.items(), key=lambda x: x[1], reverse=True):
        print(eq, pts)
```
