### Ejercicio 1 Calculadora del Índice de Masa Corporal (IMC)
Se desarrolla una función que calcula el IMC a partir del peso y la altura de una persona, además de clasificar el resultado según las categorías habituales.

```python
def calcular_imc(peso, altura):
    """
    Calcula el Índice de Masa Corporal (IMC) y devuelve
    la categoría correspondiente según el resultado.

    Parameters
    ----------
    peso : float
        Peso de la persona en kilogramos.

    altura : float
        Altura de la persona en metros.

    Returns
    -------
    tuple
        Una tupla con:
        - IMC redondeado a dos decimales.
        - Categoría del IMC.
    """

    imc = peso / (altura ** 2)

    if imc < 18.5:
        categoria = "Bajo peso"
    elif imc < 25:
        categoria = "Peso normal"
    elif imc < 30:
        categoria = "Sobrepeso"
    else:
        categoria = "Obesidad"

    return round(imc, 2), categoria


def pedir_numero(mensaje):
    """
    Solicita al usuario un número positivo y valida
    que la entrada sea correcta.

    Parameters
    ----------
    mensaje : str
        Texto que se mostrará al usuario.

    Returns
    -------
    float
        Número positivo introducido por el usuario.
    """

    while True:
        try:
            valor = float(input(mensaje).replace(",", "."))

            if valor > 0:
                return valor

            print("El valor debe ser mayor que 0.")

        except ValueError:
            print("Debes introducir un número válido.")
```

```python
peso = pedir_numero("Introduce tu peso en kilogramos: ")
altura = pedir_numero("Introduce tu altura en metros: ")

imc, categoria = calcular_imc(peso, altura)

print(f"\nTu IMC es {imc}")
print(f"Clasificación: {categoria}")
```

Existen diversas unidades de temperatura utilizadas en diferentes contextos y regiones. Las más comunes son Celsius (°C), Fahrenheit (°F) y Kelvin (K). También existen otras unidades como Rankine (°Ra) y Réaumur (°Re).
Selecciona al menos 2 conversores, de manera que al introducir una temperatura devuelva, como mínimo, dos conversiones, de forma que se puedan guardar (recuerda que un print() nunca puede guardarse).
Consejo: Intenta validar previamente los datos para que envíe un mensaje de advertencia si los datos introducidos por el usuario no tienen el formato adecuado.
(EXTRA): Piensa en una forma de almacenar todas las posibles conversiones en un único objeto (¿Lista? ¿Diccionario? ¿DataFrame?) en lugar de escribir muchos if-else según la temperatura de origen y destino.

```python
def convertir_temperatura(valor, unidad):
    """
    Convierte una temperatura entre Celsius,
    Fahrenheit y Kelvin.

    Parameters
    ----------
    valor : float
        Temperatura a convertir.

    unidad : str
        Unidad de origen ('C', 'F' o 'K').

    Returns
    -------
    dict | None
        Diccionario con las conversiones o None
        si la unidad no es válida.
    """

    unidad = unidad.upper()

    if unidad == "C":

        return {
            "Celsius": valor,
            "Fahrenheit": (valor * 9 / 5) + 32,
            "Kelvin": valor + 273.15
        }

    if unidad == "F":

        celsius = (valor - 32) * 5 / 9

        return {
            "Celsius": celsius,
            "Fahrenheit": valor,
            "Kelvin": celsius + 273.15
        }

    if unidad == "K":

        celsius = valor - 273.15

        return {
            "Celsius": celsius,
            "Fahrenheit": (celsius * 9 / 5) + 32,
            "Kelvin": valor
        }

    return None
```

### Ejercicio 3 Contador de palabras 
Escribe una función que, dado un texto, muestre cuántas veces aparece cada palabra. Intenta gestionar todas las casuísticas posibles que puedan hacer que el programa no funcione correctamente.
(EXTRA): ¿Cuál es la longitud media de las palabras del texto que has escrito?

```python
def limpiar_texto(texto):
    """
    Limpia un texto eliminando signos de puntuación
    y convirtiendo todo a minúsculas.

    Parameters
    ----------
    texto : str

    Returns
    -------
    list
        Lista de palabras limpias.
    """

    texto = texto.lower()
    texto = re.sub(r"[^a-záéíóúñü\s]", "", texto)
    return texto.split()


def contar_palabras(texto):
    """
    Cuenta la frecuencia de palabras en un texto.

    Parameters
    ----------
    texto : str

    Returns
    -------
    dict
        Diccionario con palabras y frecuencias.
    """

    palabras = limpiar_texto(texto)
    return dict(Counter(palabras))


def longitud_media_palabras(texto):
    """
    Calcula la longitud media de las palabras de un texto.

    Parameters
    ----------
    texto : str

    Returns
    -------
    float
        Longitud media de palabras.
    """

    palabras = limpiar_texto(texto)
    return sum(len(p) for p in palabras) / len(palabras)
```

```python
texto = input("Introduce un texto: ")
frecuencias = contar_palabras(texto)

print("\nFrecuencia de palabras:")
print(frecuencias)

print("\nLongitud media de palabras:")
print(round(longitud_media_palabras(texto), 2))
```

### 4. Diccionario inverso (con posibilidad de duplicados)
Resulta que el cliente tiene una encuesta muy antigua que se almacena en un diccionario y necesita los resultados al revés, es decir, intercambiando las claves y los valores.
Las claves y los valores en el diccionario original deben ser únicos. Si este no es el caso, la función deberá imprimir un mensaje de advertencia junto con una lista de los valores asociados a la clave repetida.
Observa los siguientes ejemplos para ver la salida esperada.

```python
def invertir_diccionario(diccionario):
    """
    Invierte las claves y valores de un diccionario.

    Si existen valores repetidos en el diccionario original,
    muestra una advertencia indicando las claves asociadas
    a dicho valor.

    Parameters
    ----------
    diccionario : dict
        Diccionario original.

    Returns
    -------
    dict
        Diccionario invertido.
    """

    diccionario_invertido = {}

    for clave, valor in diccionario.items():

        if valor not in diccionario_invertido:

            diccionario_invertido[valor] = clave

        else:

            if not isinstance(
                diccionario_invertido[valor],
                list
            ):

                diccionario_invertido[valor] = [
                    diccionario_invertido[valor]
                ]

            diccionario_invertido[valor].append(clave)

    for valor, claves in diccionario_invertido.items():

        if isinstance(claves, list):

            print(
                f"Advertencia: el valor '{valor}' "
                f"está repetido en las claves {claves}"
            )

    return diccionario_invertido
```

```python
datos = {
    "X": "apple",
    "Y": "banana",
    "Z": "banana",
    "F": "apple"
}

resultado = invertir_diccionario(datos)

print("\nDiccionario invertido:")
print(resultado)
```

### 1. Contador y ordenador de palabras de un texto
El cliente quedó satisfecho con el contador de palabras, pero ahora quiere leer archivos TXT y calcular la frecuencia de cada palabra, ordenadas dentro de las entradas habituales del diccionario según la letra con la que comienzan.
Es decir, las claves deben ir de la A a la Z y, dentro de cada letra, las palabras también deben ir de la A a la Z.
Por ejemplo, para el archivo: tu_me_quieres_blanca.txt 

```python
def leer_archivo(ruta):
    """
    Lee el contenido completo de un archivo.

    Parameters
    ----------
    ruta : str

    Returns
    -------
    str
    """

    with open(ruta, encoding="utf-8") as archivo:
        return archivo.read()
    

def organizar_palabras(ruta):
    """
    Organiza las palabras de un archivo TXT por
    letra inicial y cuenta sus apariciones.

    Parameters
    ----------
    ruta : str

    Returns
    -------
    dict
    """

    texto = leer_archivo(ruta)

    palabras = re.findall(
        r"\b[\wáéíóúüñ]+\b",
        texto.lower()
    )

    conteo = Counter(palabras)

    resultado = {}

    for palabra, frecuencia in conteo.items():

        letra = palabra[0]

        if letra not in resultado:
            resultado[letra] = {}

        resultado[letra][palabra] = frecuencia

    resultado_ordenado = {}

    for letra in sorted(resultado):

        resultado_ordenado[letra] = dict(
            sorted(resultado[letra].items())
        )

    return resultado_ordenado
```

### 2. Conversión de tipos de datos
El cliente recibe una lista de datos y necesita generar dos listas:
•	La primera contendrá todos los elementos que se pudieron convertir a tipo float. 
•	La segunda contendrá los elementos que no se pudieron convertir. 
Ejemplo de lista recibida:
['1.3', 'one', '1e10', 'seven', '3-1/2',
 ('2', 1, 1.4, 'not-a-number'),
 [1, 2, '3', '3.4']]
Imagen 4  
['1.3', 'one', 'le10', 'seven', '3-1/2', ('2',1,1.4,'not-a-number'), [1,2,'3','3.4']]

conversion([ '1.3', 'one' 'le10', 'seven', '3-1/2', ('2',1,1.4, 'not-a-number'), [1,2,'3', '3.4']])

([1.3, 10000000000.0, 2.0, 1.0, 1.4, 1.0, 2.0, 3.0, 3.4], ['one', 'seven', '3-1/2', 'not-a-number'])


```python
def es_float(valor):
    """
    Comprueba si un valor puede convertirse a float.

    Parameters
    ----------
    valor : any

    Returns
    -------
    bool
        True si puede convertirse a float.
    """

    try:
        float(valor)
        return True

    except (ValueError, TypeError):
        return False
    
def conversion(lista):
    """
    Separa los elementos convertibles
    y no convertibles a float.

    Parameters
    ----------
    lista : list

    Returns
    -------
    tuple
        (convertibles, no_convertibles)
    """

    convertibles = []
    no_convertibles = []

    for elemento in lista:

        if isinstance(elemento, (list, tuple)):

            for subelemento in elemento:

                if es_float(subelemento):
                    convertibles.append(float(subelemento))
                else:
                    no_convertibles.append(subelemento)

        else:

            if es_float(elemento):
                convertibles.append(float(elemento))
            else:
                no_convertibles.append(elemento)

    return convertibles, no_convertibles
```

```python
datos = [
    '1.3',
    'one',
    '1e10',
    'seven',
    '3-1/2',
    ('2', 1, 1.4, 'not-a-number'),
    [1, 2, '3', '3.4']
]

validos, invalidos = conversion(datos)

print("Convertibles:")
print(validos)

print("\nNo convertibles:")
print(invalidos)
```

### 1. Generador de contraseñas
Explora el funcionamiento del módulo random de la biblioteca NumPy.
Random module in NumPy
En este punto, el cliente ha detectado un problema con las contraseñas que utilizan sus trabajadores:
Asdf1234
fechas de cumpleaños o similares.
Para solucionarlo, nos ha encargado una función de Python que genere contraseñas más seguras.
La función debe depender de los siguientes parámetros:
longitud (int): longitud de la contraseña
mayusculas (bool = True): si deben aparecer mayúsculas
minusculas (bool = True): si deben aparecer minúsculas
numeros (bool = True): si deben aparecer números
signos (bool = False): si deben aparecer caracteres especiales (, - $ ? o similares)
Por lo tanto, si ejecutamos la función de la siguiente forma:
crear_contrasena(10, True, True, True, True)
Deberíamos obtener una salida (que debemos poder guardar) del estilo:
9Er,5Vn8P$
Asegúrate de que se cumplan todos los criterios y de que las contraseñas sean realmente aleatorias.
(EXTRA): Explora cómo podríamos hacer que la función copiara automáticamente la contraseña al portapapeles del ordenador (como si la hubiéramos seleccionado y presionado Ctrl+C).


```python
def crear_contrasena(longitud,
                     mayusculas=True,
                     minusculas=True,
                     numeros=True,
                     signos=False):
    """
    Genera una contraseña aleatoria segura usando NumPy.

    Parametros
    ----------
    longitud : int
        Longitud total de la contraseña.

    mayusculas : bool
        Incluye letras mayúsculas.

    minusculas : bool
        Incluye letras minúsculas.

    numeros : bool
        Incluye números.

    signos : bool
        Incluye caracteres especiales.

    Returns
    -------
    str
        Contraseña generada.
    """

    minus = string.ascii_lowercase
    mayus = string.ascii_uppercase
    nums = string.digits
    signos_lista = ",-$.!?@#%&*"

    caracteres = []

    if minusculas:
        caracteres.extend(minus)

    if mayusculas:
        caracteres.extend(mayus)

    if numeros:
        caracteres.extend(nums)

    if signos:
        caracteres.extend(signos_lista)

    if len(caracteres) == 0:
        return "Error: no hay caracteres disponibles"

    contrasena = np.random.choice(caracteres, size=longitud)

    return "".join(contrasena)
```

```python
password = crear_contrasena(10, True, True, True, True)

print("Contraseña generada:")
print(password)
```

### 2. Procesamiento simple de datos
Una compañera de trabajo nos ha pedido un favor, aprovechando que sabe que estamos aprendiendo a programar.
Tiene un histórico de partidos de fútbol catalán en un archivo donde se almacenan los nombres de los equipos y los resultados. Necesita que procesemos los datos de forma automática para extraer la información que necesita.
Utiliza el archivo:
historic_partits.txt
Necesita un programa que devuelva:
•	El número total de goles marcados por cada equipo. 
•	El nombre del equipo más goleador. 
•	El nombre del equipo más goleado. 
•	La clasificación global (victoria = 3 puntos, empate = 1 punto, derrota = 0 puntos). 
Como referencia, los equipos con más puntos serían los siguientes:
Imagen resultado partidos
Puedes utilizar el formato que te resulte más conveniente.


```python
def leer_partidos(ruta):
    """
    Lee el archivo de partidos y lo convierte en una lista estructurada.

    Parameters
    ----------
    ruta : str
        Ruta del archivo de texto.

    Returns
    -------
    list
        Lista de tuplas con formato:
        (equipo_local, goles_local, goles_visitante, equipo_visitante)
    """

    partidos = []

    with open(ruta, "r", encoding="utf-8") as file:

        for linea in file:
            linea = linea.strip()

            if not linea:
                continue

            local, marcador, visitante = linea.split("\t")
            goles_local, goles_visitante = marcador.split("-")

            partidos.append(
                (local,
                 int(goles_local),
                 int(goles_visitante),
                 visitante)
            )

    return partidos
    
def procesar_estadisticas(partidos):
    """
    Calcula estadísticas de goles y puntos por equipo.

    Parameters
    ----------
    partidos : list

    Returns
    -------
    tuple
        (goles_favor, goles_contra, puntos)
    """

    goles_favor = defaultdict(int)
    goles_contra = defaultdict(int)
    puntos = defaultdict(int)

    for local, gl, gv, visitante in partidos:

        # goles a favor y en contra
        goles_favor[local] += gl
        goles_favor[visitante] += gv

        goles_contra[local] += gv
        goles_contra[visitante] += gl

        # puntos
        if gl > gv:
            puntos[local] += 3
        elif gl < gv:
            puntos[visitante] += 3
        else:
            puntos[local] += 1
            puntos[visitante] += 1

    return goles_favor, goles_contra, puntos

def equipo_maximo(diccionario):
    """
    Devuelve la clave con el valor máximo de un diccionario.

    Parameters
    ----------
    diccionario : dict

    Returns
    -------
    str
        Equipo con mayor valor.
    """

    return max(diccionario, key=diccionario.get)
```

```python
ruta = "historic partits.txt"

partidos = leer_partidos(ruta)

goles_favor, goles_contra, puntos = procesar_estadisticas(partidos)

# Resultados
print("\nGoles a favor por equipo:")
print(dict(goles_favor))

print("\nEquipo más goleador:")
print(equipo_maximo(goles_favor))

print("\nEquipo más goleado:")
print(equipo_maximo(goles_contra))

print("\nClasificación (Top 10):")
clasificacion = sorted(puntos.items(), key=lambda x: x[1], reverse=True)

for equipo, pts in clasificacion[:10]:
    print(f"{equipo}: {pts} puntos")
```
